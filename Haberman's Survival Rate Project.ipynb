```python
#Univaraite Analysis
```

```python
#Boxplot

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(x="Survival_Status", y="Age", data=haberman, ax=axes[0])
axes[0].set_title("Age")


sns.boxplot(x="Survival_Status", y="Year", data=haberman, ax=axes[1])
axes[1].set_title("Year")


sns.boxplot(x="Survival_Status", y="Nodes", data=haberman, ax=axes[2])
axes[2].set_title("Nodes")


plt.tight_layout()


plt.show()



```

```python
#Violin Plots
sns.violinplot(x="Survival_Status", y="Age", data=haberman, size=3)
plt.show()

sns.violinplot(x="Survival_Status", y="Year", data=haberman, size=3)
plt.show()

sns.violinplot(x="Survival_Status", y="Nodes", data=haberman, size=3)
plt.show()
```

```python
#Bi-variate analysis
#Pair plots
sns.pairplot(haberman, hue="Survival_Status", height=4)
plt.show()
```

```python
#Scatter Plot
sns.scatterplot (x='Age' , y= 'Nodes', hue='Survival_Status', data=haberman)
plt.show()
```

```python
# Observations:
1. We can see the distribution of specific features for various survival statuses using boxplots and violin plots.
2. Pair plots assist visualize relationships between features by displaying scatter plots for each pairing of features.
```
